```python
# A university monitors electricity usage of 4 hosteks over 7 days (mon to sun).
 
import numpy as np
```

Task A

```python
#Task A1: Create a Numpy array electricity using np.random.randint.range(values between 120 and 650),(shape 4,7) and print the array.

electricity=np.random.randint(120,650,size=(4,7))
print("Electricity usage array:\n", electricity)
```

```python
#Task A2: Print Shape, Number of dimensions, Total number of elements and Data type of the array.

print("Shape of the array:", electricity.shape)
print("Number of dimensions:", electricity.ndim)
print("Total number of elements:", electricity.size)
print("Data type of the array:", electricity.dtype)
```

Task B

```python
#Task B1: Find the average electricity usage per hostel, the maximum usage per hostel an identify which hostel has the highest average usage.

average_usage_per_hostel = np.mean(electricity, axis=1)
max_usage_per_hostel = np.max(electricity, axis=1)
highest_average_usage_hostel = np.argmax(average_usage_per_hostel)
print("Average electricity usage per hostel:", average_usage_per_hostel)
print("Maximum electricity usage per hostel:", max_usage_per_hostel)
print("Hostel with the highest average usage:", highest_average_usage_hostel)
```

```python
#Task B2: Calculate the total electricity consumption for each day and identify which day has the highest consumption.

total_consumption_per_day = np.sum(electricity, axis=0)
highest_consumption_day = np.argmax(total_consumption_per_day)
print("Total electricity consumption per day:", total_consumption_per_day)
print("Day with the highest consumption (0=Mon, 1=Tue, ..., 6=Sun):", highest_consumption_day)
```

Task C

```python
#Task C1: Create a new array penalized usage such that
    #- Values > 550 -> multiply by 1.2
    #- All other values remain the same.

penalized_usage = np.where(electricity > 550, electricity * 1.2, electricity)
print("Penalized electricity usage array:\n", penalized_usage)
```

Task D

```python
#Task D1: Compare each hostel's daily usage with its own weekly average. Create a boolean matrix where:
    #- True: Above hostel average
    #- False: Below or equal to hostel average

hostel_average = np.mean(electricity, axis=1, keepdims=True)
above_average_matrix = electricity > hostel_average
print("Boolean matrix indicating above average usage:\n", above_average_matrix)
```

Task E

```python
#Task E1: Reshape the electricity data into single row representing weekly usage (shape 1,28) and print the reshaped array.

reshaped_weekly_usage = electricity.reshape(1, -1)
print("Reshaped weekly usage array (1,28):\n", reshaped_weekly_usage)
```

```python
#Task E2: Reshape the same data into 2 rows and 14 columns (shape 2,14). Ensure no data is lost or duplicated. Print the reshaped array.

reshaped_2_rows = electricity.reshape(2, 14)
print("Reshaped array (2,14):\n", reshaped_2_rows)
```

Task F

```python
#Task F1: Create ref_backup = electricty
    # Create safe_backup = ref_backup.copy()
    # Modify one value in the original electricity array.
    # Observe which backups change.

ref_backup = electricity
safe_backup = ref_backup.copy()
electricity[0, 0] = 999  # Modify one value in the original array
print("Original electricity array after modification:\n", electricity)
print("Reference backup (ref_backup):\n", ref_backup)
print("Safe backup (safe_backup):\n", safe_backup)
 
```
