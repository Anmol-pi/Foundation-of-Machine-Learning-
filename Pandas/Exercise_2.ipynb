```python
import pandas as pd
```

```python
df = pd.DataFrame({
    'Name': ['Alice Johnson', 'Bob Smith', 'Charlie Brown', 'Diana Prince', 'Eve Wilson', 'Frank Miller', 'Grace Lee', 'Henry Davis', 'Iris Martinez', 'Jack Taylor'],
    'Department': ['IT', 'HR', 'Sales', 'IT', 'Finance', 'Sales', 'HR', 'Finance', 'IT', 'Sales'],
    'Role': ['Developer', 'Manager', 'Analyst', 'Manager', 'Analyst', 'Executive', 'Specialist', 'Coordinator', 'Developer', 'Rep'],
    'Experience': [5, 8, None, 10, 9, 12, 4, 2, 7, 5],
    'Salary': [75000, 85000, 65000, 95000, None, 120000, 68000, 55000, 80000, None],
    'Score': [8.5, 9.0, 7.5, 8.8, 8.2, 9.5, None, 7.0, 8.9, 8.1]
})
df
```

Task 1

```python

# What is the average(mean) Salary for each Department?
print("Average Salary for each Department:")
df.groupby('Department')['Salary'].mean()
```

```python
# What is the average(mean) Performance Score for each Role?
print("Average Performance Score for each Role:")
df.groupby('Role')['Score'].mean()
```

```python
# How many actual, non-null Performance Score entries exist for each Department?
print("Non-null Performance Score entries for each Department:")
df.groupby('Department')['Score'].count()
```

```python
# What is the total number of employees records(rows) in each Department regardless of whether data is missing?
print("Total number of employees records for each Department:")
df.groupby('Department').size()
```

Task 2


```python
# Generate a summary table for Salary based on Department that shows the mean, maximum, minimum, and count in a single line of code.
df.groupby('Department')['Salary'].agg(['mean', 'max', 'min', 'count'])
```

```python
# Generate the same summary table (mean, max, min, count) but analyze the Performance Score based on the Role
df.groupby('Role')['Score'].agg(['mean', 'max', 'min', 'count'])
```

```python
# Find the average Salary grouped by both Department and Role
df.groupby(['Department', 'Role'])['Salary'].mean()
```

```python
# Find the average Performance Score grouped by both Department and Experience level
df.groupby(['Department', 'Experience'])['Score'].mean()
```
